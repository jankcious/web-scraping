```python
from selenium import webdriver
from bs4 import BeautifulSoup

store_list = []
browser = webdriver.Firefox()
for URL in Full_Clean_URLs:
    browser.get(URL)
    content = browser.page_source
    soup = BeautifulSoup(content, 'lxml')
    address = soup.find_all(attrs={'class':'store-address1'})
    address2 = soup.find_all(attrs={'class':'store-address2'})
    for store in address:
        store_list.append([store.contents[0].strip()])
    for store in address2:
        store_list[-1].append(store.contents[0].strip())
browser.close()
with open('Sephora_Locations.csv', 'wt', newline='') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerows(store_list)
```
